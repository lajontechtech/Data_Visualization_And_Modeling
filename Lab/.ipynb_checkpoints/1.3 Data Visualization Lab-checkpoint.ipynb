```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1')
df5 = df.head(5)
df.head()
```

```python
df.columns

```

```python
df5_bpm = df.head(5).sort_values(by='bpm')
plt.figure(figsize=(10,4))
plt.bar(df5_bpm['track_name'], df5_bpm['bpm'])
plt.title('Distribution of BPM')
plt.xlabel('Song')
plt.ylabel('BPM')
plt.xticks(rotation=45, ha='right')
plt.tight_layout
plt.show()
```

BPM Description:
This graph shows the BPM, or tempo, of the first 5 songs. The songs range from a low of about 92 BPM for LALA to a high of about 170 BPM for Cruel Summer. Most of the songs are in the middle range between about 125 and 145 BPM.

```python
df5_dance = df.head(5).sort_values(by='danceability_%')
plt.figure(figsize=(10,4))
plt.bar(df5_dance['track_name'], df5_dance['danceability_%'])
plt.title('Danceability %')
plt.xlabel('Song')
plt.ylabel('Danceability (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout
plt.show()
```

Danceability description:
This graph shows how danceable the first 5 songs are. The values go from about 51% to 80%. Seven has the highest danceability, while vampire has the lowest, so the songs vary quite a bit in how easy they are to dance to.

```python
df5['key'].value_counts().plot(kind='bar', figsize=(6,4))
plt.title('Key Counts')
plt.xlabel('Key')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

Key description:
This graph shows how many songs fall into each musical key. In this sample, A appears the most with 2 songs, while B, C#, and F each appear 1 time. This means A is the most common key in the 5-song sample.

```python
df5_mode = df.head(5).sort_values(by='mode')
plt.figure(figsize=(10,4))
plt.bar(df5_mode['track_name'], df5_mode['mode'])
plt.title('Mode')
plt.xlabel('Song')
plt.ylabel('Mode')
plt.xticks(rotation=45, ha='right')
plt.tight_layout
plt.show()
```

Mode description:
This graph shows the musical mode of the songs, which is either major or minor. Based on the graph, minor appears to be the common mode in this sample. That means most of these songs seem to use a minor sound instead of a major one.

```python
df5_energy = df.head(5).sort_values(by='energy_%')
plt.figure(figsize=(10,4))
plt.bar(df5_energy['track_name'], df5_energy['energy_%'])
plt.title('Distribution of Energy')
plt.xlabel('Song')
plt.ylabel('Energy')
plt.xticks(rotation=45, ha='right')
plt.tight_layout
plt.show()
```

Energy description:
This graph shows the energy level of the songs. The values range from about 53 to 83. Most of the songs are on the higher side, which suggests the sample mostly contains songs with a strong and lively sound.

```python
df5_acouticness = df.head(5).sort_values(by='acousticness_%')
plt.figure(figsize=(10,4))
plt.bar(df5_acouticness['track_name'], df5_acouticness['acousticness_%'])
plt.title('Distribution of Acousticness')
plt.xlabel('Song')
plt.ylabel('Acousticness(%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout
plt.show()
```

Acousticness description:
This graph shows how acoustic each song is. The values range from about 7% to 31%. Most of the songs are on the lower side, which means they are more produced and less acoustic.

```python
df5_speech = df.head(5).sort_values(by='speechiness_%')
plt.figure(figsize=(10,4))
plt.bar(df5_speech['track_name'], df5_speech['speechiness_%'])
plt.title('Distribution of Speechiness')
plt.xlabel('Song')
plt.ylabel('Speechiness')
plt.xticks(rotation=45, ha='right')
plt.tight_layout
plt.show()
```

Speechiness Description:
This graph shows how much spoken-word style is in each song. The values range from about 4 to 15. Most songs are low in speechiness, but Cruel Summer stands out as having the highest value in this sample.
